```python
import numpy as np, matplotlib.pyplot as plt 
```

```python
N, M = 1000, 10
(x, y), label = np.random.randn(2, N), np.random.randint(0, M, N)

jet = plt.cm.get_cmap("jet")
colors = jet(np.linspace(0, 1, N))
cmap = jet.from_list(f"jet{N}", colors, N)
fig, axes = plt.subplots(1, 2)
fig.suptitle("scatter with 10 discrete color")
axes[0].scatter(x, y, 15, label, None, "jet")
im = axes[1].scatter(x, y, 15, label, 'o', cmap, edgecolor='none')
plt.colorbar( im, ticks = range(10) )
None
```
